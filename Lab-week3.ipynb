### Installing Git
* We will follow the steps indicated through the software-carpentry lesson on *Version Control with Git* in the *Setup*.

### Creating a new repository on GitHub

* Create a GitHub account if you do not already have one.
* Go to your GitHub profile page
* Click on **Repositories**
* Click on **New**
* Give your new repository a name like "E11Work" or something similar
* Under *Initialize your repository with:*
  * Check *Add a README file*
  * Check *Add .gitignore*
* Click **Create Repository**
* Go to *Settings*
* Click on **Collaborators** on the left side-bar (you will be prompted for your password)
* Add me (alihanks)and Chris (cllamb0) as collaborators so that we can view your homework submissions.

### Creating a local clone of your repository

* From your the main page of your new repository, click on the **↓ Code** dropdown
* Select **Open with GitHub Desktop** 
  * NOTE: If you are already comfortable with git, GitHub and the command line methods, you may procede in the usual way.
* Download the installation executable for the GitHub Desktop App and install the App.
* Run the app and login to your GitHub account through the app.
* Select to *Clone a repository from the Internet...* and select the repository you just created from the list of options in the pop-up.

### Putting your homework into your repository

* Open your **File Explorer** or **File Manager** or similar.
* Navigate to the folder currently storing your homework
* Open another **File Explorer** window and navigate to your new GitHub repo area
  * NOTE: The GitHub Desktop app will have created a **GitHub** folder in your **Documents** folder and cloned your repository there.
* Copy or drag and drop your homework notebook and any other notebooks you've created for this class into your repository folder.
* On the GitHub Desktop app you should now see that changes have been made to your local version of your repository that need to be committed and pushed to the version of your repository on GitHub. 
  * Click on **Commit to master**
    * NOTE: add a comment that describes the changes, in this case something like "first homework submission" or something similar
  * Click on **Push origin**
* All of your work should now show up on the GitHub page for your repository. Your homework has been submitted.
* The graded version of your homework will be returned to you through GitHub as well as an update to your notebook.
