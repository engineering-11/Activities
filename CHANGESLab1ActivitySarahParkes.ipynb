   #### 1. Import the libraries needed for reading in and manipulating data

```python
import numpy as np
import pandas as pd
# This line allows plots to show in the Jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt
```

#### 2. Import data using pandas

```python
data = pd.read_csv('https://radwatch.berkeley.edu/sites/default/files/dosenet/etch_roof.csv')
```

##### 2.a Take a look at what the data looks like
The data has been loaded into a pandas DataFrame, a fancy multi-dimensional list, basically.

```python
data
```

```python
data.describe()
```

#### 3. Slice/cut the dataframe so that only the local time, cpm and cmpError are stored

```python
cpm_data = data.loc[:,"deviceTime_local":"cpmError"]
cpm_data = cpm_data.drop(["deviceTime_unix"],axis=1)
cpm_data
```

#### 4. Find the max and min values in the DataFrame

```python
print(cpm_data.max())
print(cpm_data.min())
```

#### 5. There aren't really times when there are zero counts, this is an indication that the detector wasn't working. Remove those entries so that we can look only at valid data.

```python
cpm_data.loc[:,"cpm"] > 0.0
mask = cpm_data["cpm"] > 0.0
#print(mask)
good_cpm_data = cpm_data[mask]
good_cpm_data.min()
```

#### 6. We also don't really need to look at all the data. Cut out all the data from before August 5th.

```python
time_mask = good_cpm_data['deviceTime_local']>'2019-08-05 00:00:00-07:00'
month_cpm = good_cpm_data[time_mask]
month_cpm
```

#### 7. Plot the cpm data using matplotlib

```python
plt.plot(good_cpm_data["cpm"])
plt.ylabel('Counts Per Minute')                       # label the y-axis
plt.title("CPM vs index")                                  # put a title!
plt.show()
```

##### 7. a. OK, but a more meaningful way to plot this is probably as a histogram, so plot this data as a histogram

```python
plt.hist(month_cpm["cpm"],bins=31)
plt.ylabel('Frequency')
plt.xlabel('Counts Per Minute')
plt.show()
```

#### 8. What is the mean CPM? What about the standard deviasion?
**Hint:** You can do this by hand, but numpy is your friend.

```python
#What is the mean CPM? What about the standard deviation?
cpm_mean = np.mean(month_cpm)
print ('the cpm mean is ') 
print (cpm_mean) #first value printed is CPM mean for month data 
cpm_standev = np.std(month_cpm)
print ('the cpm standard deviation is ')
print (cpm_standev) #first value printed is CPM standard deviation for month data
```
